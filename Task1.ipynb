```python
# Null values percentage per column
null_percentage = data.isnull().sum() / len(data) * 100

# Filter to show only columns with null values and sort them
null_percentage = null_percentage[null_percentage > 0].sort_values(ascending=False)

if null_percentage.empty:
    print("No null values found in any column.")
else:
    print(null_percentage)
```

```python
# percentage of null vales is <5% so we can drop the null values.
# Droping null values
data=data.dropna()
```
